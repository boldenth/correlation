```python

test_dates = ["2017-11-19", "11/19/2017", "19 Nov 2017"]

def convert_time(time):
    
    r_str = ""
    
    if time.find('-') != -1:
        
        r_str = str(dt.datetime.strptime(time, '%Y-%m-%d'))\
                    .replace(' ', 'T') \
                    .replace(':','')   \
                    .replace('-','')
    
    elif time.find('/') != -1:
        
        r_str = str(dt.datetime.strptime(time, '%m/%d/%Y'))\
                    .replace(' ', 'T') \
                    .replace(':','')   \
                    .replace('-','')
    
    else:
        
        abbr_to_num = {name.lower(): num for num, name \
                      in enumerate(cal.month_abbr) if num}

        name_to_num = {name.lower(): num for num, name \
                      in enumerate(cal.month_name) if num}
        
        try:
            
            three = time.split()
            
            try:
                
                three = [int(three[0]), 
                         abbr_to_num[three[1].lower()], 
                         int(three[2])]
                         
            except:
                three = [three[0], name_to_num[three[1].lower()], three[2]]
            
            r_str = str(dt.datetime(year=three[2],
                                    month=three[1],
                                    day=three[0]))\
                    .replace(' ', 'T') \
                    .replace(':','')   \
                    .replace('-','')
            
        except:
            print("Fail")
            r_str = time
    
    return r_str

for s in test_dates:
    print(convert_time(s))
```

```python
df = pd.read_csv("test/SSA_births_2000-2014.csv")

dropping = (1000,2500)

df = df.drop(df.index[:dropping[0]]) \
       .drop(df.index[dropping[-1]:])\
       .reset_index(drop=True)
        
df
```

```python
def sum_years(dee_eff, date_col=None, colname=None):
    
    if colname is None:
        
        return
    
    to_avg = df[colname]
    
    new_df = pd.DataFrame()
    
    temp_dict = {} # {"year":colname} :: {'year': [year,]}
    
    for index, row in dee_eff.iterrows():
        
        year = str(row[date_col])[:4]
        
        if year in temp_dict:
            temp_dict[year] += row[colname]
            
        else:
            temp_dict[year] = row[colname]
    
    #print(temp_dict)
    
    temp_list = []
    
    #for key in sorted(temp_dict.iterkeys()):
    #    temp_list.append[]
    
    new_df = pd.DataFrame.from_dict( temp_dict, orient='index')#, columns=["year","births"])

    
    new_df_b = pd.DataFrame(list(temp_dict.items()), 
                            columns=["year",colname])

    #print(new_df)
    #print(new_df_b)
    
    return new_df_b
    
fraym = sum_years(df, date_col='year', colname='births')
```

```python
def normalize_col(frame, col_name=None, zero=False):
    
    temp_col = []
    
    low, high = 1062688, 4211941
    
    if zero == False:
        for index, row in frame.iterrows():
            temp_col.append(frame[col_name][index]/high)
    
    else:
        for index, row in frame.iterrows(): # this-min / max-min
            temp_col.append((frame[col_name][index] - low) / (high - low))
        
    frame["normalized " + col_name] = temp_col
        
normalize_col(fraym, col_name="births", zero=True)

fraym
```
